# Stack array-based

```python
class stack_arr:
    def __init__(self,size):
        self.size=size
        self.stack=[]
        self.top=0
    def push(self,data):
        pushed =False
        if self.top<self.size:
            self.stack.append(data)
            self.top=self.top+1
            pushed=True
        return pushed
    def pop(self):
        retval=-1
        if self.top>0:
            self.top=self.top-1
            retval=self.stack[self.top]
        return retval
    def print_stack(self):
        print('the stack is:')
        for i in range(self.top-1,-1,-1):
            print(self.stack[i])

s=stack_arr(5)
is_pushed=s.push(10)
if is_pushed:
    print("the value is pushed successfully")
s.print_stack()
is_pushed=s.push(20) 
if is_pushed:
    print("the value is pushed successfully")  
s.print_stack()
is_pushed=s.push(30)    
if is_pushed:
    print("the value is pushed successfully")
s.print_stack()
print("Popped:",s.pop())
s.print_stack()

```

# Queue

```python
class queue_arr:
    def __init__(self, size):
        self.size = size
        self.queue = [0] * size  
        self.rear = 0            
    def enqueue(self, data):
        if self.rear < self.size:
            self.queue[self.rear] = data
            self.rear += 1
            return True
        return False  
    def dequeue(self):
        if self.rear == 0:   
            return -1
        retval = self.queue[0]  
        for i in range(1, self.rear):
            self.queue[i-1] = self.queue[i]
        self.rear -= 1  
        return retval
    def print_queue(self):
        print("the queue is:")
        if self.rear == 0:
            print("empty")
            return
        for i in range(self.rear):
            print(self.queue[i])

q=queue_arr(5)
is_enqueued=q.enqueue(10)   
if is_enqueued:
    print("the value is enqueued successfully")
q.print_queue()
is_enqueued=q.enqueue(20)
if is_enqueued:
    print("the value is enqueued successfully")
q.print_queue()
is_enqueued=q.enqueue(30)
if is_enqueued:
    print("the value is enqueued successfully")
q.print_queue()
is_enqueued=q.enqueue(40)
if is_enqueued:
    print("the value is enqueued successfully")
q.print_queue()
is_enqueued=q.enqueue(50)
if is_enqueued:
    print("the value is enqueued successfully")
q.print_queue()
is_enqueued=q.enqueue(60)
if is_enqueued:
    print("the value is enqueued successfully")
else:
    print("the queue is full")
q.print_queue()
print("Dequeued:",q.dequeue())
q.print_queue()
is_enqueued=q.enqueue(80)   
if is_enqueued:
    print("the value is enqueued successfully")
else:
    print("the queue is full")
q.print_queue()
```
